```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q=SELECT'
import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

### Task 3 — Working with Space
###### Showing the list of start station locations
###### Using GROUP BY
###### Showing the number of trips started per station
###### but only for stations within 500m of Time Square!
###### The coordinates of Time Square is (40.7577,-73.9857)

```python
# Showing the list of start station locations
question3_1 = '''
select start_station_id,start_station_longitude, start_station_latitude
from citibike'''
pd.read_csv(StringIO.StringIO(queryCartoDB(question3_1)), sep=',')
```

```python
# Showing the number of trips started per station
question3_2 = '''
select count(tripduration),start_station_id
from citibike
group by start_station_id
'''
pd.read_csv(StringIO.StringIO(queryCartoDB(question3_2)), sep=',')
```

```python
# for stations within 500m of Time Square!
# The coordinates of Time Square is (40.7577,-73.9857)
question3_3 = '''
select count(tripduration),start_station_id
from citibike
where st_dwithin(
CDB_LatLng(start_station_latitude, start_station_longitude)::geography, CDB_LatLng(40.7577, -73.9857)::geography, 500)
Group by start_station_id
'''
pd.read_csv(StringIO.StringIO(queryCartoDB(question3_3)), sep=',')
```

### Task 4 — Putting it all together
###### Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare
###### Extra create lines for trips started from stations within 500m of Times
###### Squares and lasted less than 2 hours. The number of trips per each pair of stations are output as attributes of these lines.

```python
#Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare
question4_1 = '''
select AVG(tripduration)as duration,start_station_name
from citibike
where st_dwithin(
CDB_LatLng(start_station_latitude, start_station_longitude)::geography, CDB_LatLng(40.7577, -73.9857)::geography, 500)
and EXTRACT(dow from starttime) in (0,6)
Group by start_station_name
Order by duration DESC
Limit 1
'''
pd.read_csv(StringIO.StringIO(queryCartoDB(question4_1)), sep=',')
```

```python
# extra credit
question4_2='''
select count(start_station_id) as num_trips, start_station_name, end_station_name
from citibike
where st_dwithin(
CDB_LatLng(start_station_latitude, start_station_longitude)::geography, CDB_LatLng(40.7577, -73.9857)::geography, 500)
and tripduration <= 7200
Group by start_station_name, end_station_name

'''
pd.read_csv(StringIO.StringIO(queryCartoDB(question4_2)), sep=',')
```
