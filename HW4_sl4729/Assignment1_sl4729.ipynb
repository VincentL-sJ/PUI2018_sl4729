### Gaussian distribution

```python
from __future__ import print_function
__author__= 'fbb'

import pylab as pl
import seaborn
import numpy as np
from scipy.optimize import curve_fit, minimize
    
%pylab inline
```

```python
#distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
distributions = ['gaus']
mymean = 101
df=mymean
```

```python
myDistribution = {}
np.random.seed(3)
myDistribution['gaus'] = np.random.normal(df, size=100)
pl.hist(myDistribution['gaus'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

Figure 1: Normal distuibution: df=101, N=100

```python
print ("Gaussian mean: %.2f, standard deviation: %.2f"%(myDistribution['gaus'].mean(), myDistribution['gaus'].std()))
```

```python
mysize = (2000 / (np.array(range(1, 101)))).astype(int)
print (mysize, mysize.shape)
```

```python
myDistribution['gaus'] = {} 

for n in mysize:
    myDistribution['gaus'][n] = np.random.normal(df, size = n)
    
myDistribution['gaus']['means'] = {}
axnormal_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in myDistribution['gaus']:
    if not type(nn) == str:
        myDistribution['gaus']['means'][nn] = myDistribution['gaus'][nn].mean()
        axnormal_mu_n.plot(nn, myDistribution['gaus']['means'][nn], 'o')
        axnormal_mu_n.set_xlabel('sample size', fontsize=18)
        axnormal_mu_n.set_ylabel('sample mean', fontsize=18)
        axnormal_mu_n.set_title('Gaussian Distribution', fontsize=18)
        axnormal_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
```

Figure 2: this graph displays the sample mean versus the sample size. As the sample size getting large, the sample mean is getting close to the population mean.

```python
samplemeans = list(myDistribution['gaus']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(samplemeans,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

Figure 3: this figure displays the histogram of the mean of all samples

### chi sq distribution

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
print (mysize, mysize.shape)
```
