```python
dist_container = []
for i in range(100):
    size=int(np.random.rand()*1900+10)
    dist=np.random.randn(size)
    dist_container.append(dist)
dist_container
```

```python
len(dist_container)
```

```python
ax = pl.figure().add_subplot(111)
for d in dist_container:
    print(d.mean(), len(d))
    ax.plot(len(d),d.mean(),'.')
ax.plot([0,2000],[0,0],'-')
```
